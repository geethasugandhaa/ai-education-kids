```python
import tkinter as tk
from tkinter import messagebox
root = tk.Tk()
root.title("Smart Kids Learning Game")
root.geometry("700x500")
root.configure(bg="#D6F5FF")
quiz = {
    "Math": [
        ("9 + 5 = ?", "14"),
        ("8 + 3 = ?", "11"),
        ("7 + 6 = ?", "13"),
        ("10 + 2 = ?", "12")
    ],
    "GK": [
        ("Capital of India?", "Delhi"),
        ("National Animal?", "Tiger"),
        ("Red Planet?", "Mars"),
        ("How many days in a week?", "7")
    ],
    "Spelling": [
        ("Spell baby dog", "Puppy"),
        ("Spell yellow fruit", "Banana"),
        ("Spell king of jungle", "Lion"),
        ("Spell flying bird", "Parrot")
    ]
}
current = ""
index = 0
score = 0
def start_quiz(category):
    global current, index, score
    current = category
    index = 0
    score = 0
    show_question()
def show_question():
    question.config(text=quiz[current][index][0])
    answer.delete(0, tk.END)
def check_answer(event=None):
    global index, score
    user = answer.get().strip().lower()
    correct = quiz[current][index][1]
    if user == correct.lower():
        score += 1
        messagebox.showinfo(
            "Result",
            f"Correct!\nAnswer: {correct}"
        )
    else:
        messagebox.showinfo(
            "Result",
            f"Wrong!\nCorrect Answer: {correct}"
        )
    index += 1
    if index < 4:
        show_question()
    else:
        messagebox.showinfo(
            "Quiz Finished",
            f" Your Final Score: {score}/4"
        )
        question.config(text="Choose Another Quiz")
        answer.delete(0, tk.END)
title = tk.Label(
    root,
    text=" AI Educational Game For Kids",
    font=("Arial", 22, "bold"),
    fg="blue",
    bg="#D6F5FF"
)
title.pack(pady=20)
question = tk.Label(
    root,
    text="Select a Quiz Category",
    font=("Arial", 18, "bold"),
    bg="#D6F5FF",
    fg="#333"
)
question.pack(pady=20)
answer = tk.Entry(
    root,
    font=("Arial", 16),
    width=25,
    justify="center"
)
answer.pack(pady=15)
answer.bind("<Return>", check_answer)
frame = tk.Frame(root, bg="#D6F5FF")
frame.pack(pady=20)
tk.Button(
    frame,
    text="➕ Math Quiz",
    bg="#FFD54F",
    font=("Arial", 12, "bold"),
    width=15,
    command=lambda: start_quiz("Math")
).grid(row=0, column=0, padx=10)
tk.Button(
    frame,
    text="🌍 GK Quiz",
    bg="#A5D6A7",
    font=("Arial", 12, "bold"),
    width=15,
    command=lambda: start_quiz("GK")
).grid(row=0, column=1, padx=10)
tk.Button(
    frame,
    text="📖 Spelling Game",
    bg="#CE93D8",
    font=("Arial", 12, "bold"),
    width=15,
    command=lambda: start_quiz("Spelling")
).grid(row=0, column=2, padx=10)
tk.Button(
    root,
    text="Submit Answer",
    bg="#4FC3F7",
    fg="white",
    font=("Arial", 14, "bold"),
    width=18,
    command=check_answer
).pack(pady=30)
footer = tk.Label(
    root,
    text=" Learn • Play • Grow",
    font=("Arial", 12, "italic"),
    bg="#D6F5FF",
    fg="darkblue"
)
footer.pack(side="bottom", pady=15)

root.mainloop()
```
